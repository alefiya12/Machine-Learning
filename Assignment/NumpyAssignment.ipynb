```python
# 1. Create a NumPy array of integers from 1 to 10 (inclusive).
arr = np.array(np.arange(1,11))
arr
```

```python
# 2. Reshape the array from the previous question into a 2x5 matrix.
arr1 = arr.reshape(2,5)
arr1
```

```python
# 3. Perform element-wise multiplication of two NumPy arrays:
mul1 = np.arange(1,12,2).reshape(3,2)
mul2 = np.arange(1,7).reshape(3,2)
print(mul1)
print()
print(mul2)
print()
print(mul1*mul2)
```

```python
# 4. Create one multidimensional array of shape (3,4,2) and perform indexing and slicing on it.
multi_arr = np.arange(24).reshape(3, 4, 2)
print("\n3D array (3,4,2):\n", multi_arr)
print("\nElement at [1, 2, 0]:", multi_arr[1, 2, 0])
print("\nSlice [0, :, :]:\n", multi_arr[0, :, :])
```

```python
# 5. Generate a 3x3 matrix with random values between 0 and 1.
random_matrix = np.random.rand(3, 3)
print("\n3x3 Random Matrix:\n", random_matrix)
```

```python
# 6. Create two 2d array and perform hstack and vstack
x = np.array([[1, 2], [3, 4]])
y = np.array([[5, 6], [7, 8]])

h_stacked = np.hstack((x, y))
v_stacked = np.vstack((x, y))
print("\nHorizontal Stack:\n", h_stacked)
print("\nVertical Stack:\n", v_stacked)
```
