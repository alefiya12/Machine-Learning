Matplotlib → low-level, highly customizable plots

Seaborn → built on top of Matplotlib, easier for statistical visualization

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 1. Create a scatter plot using Matplotlib with random data points. Customize the markers and colors
x = np.random.randint(1,100,50)
y = np.random.randint(1,100,50)

plt.scatter(x, y, s=50)
plt.title("Scatter Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

```python
# 2. Create a bar chart to visualize the sales data for different products.(Take the sales data from Kaggle)
sales = pd.read_excel("/content/Coffee Shop Sales.xlsx")

plt.bar(sales['product_id'], sales['unit_price'], color="orange")
plt.title("Sales Data for Products")
plt.xlabel("Products")
plt.ylabel("Unit Price")
plt.show()
```

```python
# 3. Plot a pie chart to represent the percentage distribution of expenses in a budget.
# Data for 3rd que:-
# categories = ['Rent', 'Food', 'Utilities', 'Entertainment']
# expenses = [1000, 300, 150, 200].
categories = ['Rent', 'Food', 'Utilities', 'Entertainment']
expenses = [1000, 300, 150, 200]

plt.pie(expenses, labels=categories, autopct='%1.1f%%', startangle=90, colors=['gold', 'lightgreen', 'lightblue', 'pink'])
plt.title("Budget Distribution")
plt.show()
```

```python
# 4. Plot a simple line graph using Matplotlib with the following data:
# x=[1,2,3,4,5]
# y=[10,15,7,12,8]
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 12, 8]

plt.plot(x, y, marker='o', linestyle='--', color='purple')
plt.title("Simple Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# 5. How do you create subplots in Matplotlib? Explain with code?
# Syntax: plt.subplots(rows, cols) creates multiple plots in one figure.
# Subplot creates a grid of plots i.e. 2 plots in 1 figure

fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# First subplot
axs[0].plot([1, 2, 3], [4, 5, 6], color='blue')
axs[0].set_title("Line Plot")

# Second subplot
axs[1].bar(['A', 'B', 'C'], [3, 7, 5], color='orange')
axs[1].set_title("Bar Chart")

plt.suptitle("Example of Subplots")
plt.show()
```

```python
# 6. How can you add labels to the x-axis and y-axis in a Matplotlib plot? Explain with code?
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x, y, color='green')
plt.xlabel("Time (hours)")     # X-axis label
plt.ylabel("Distance (km)")    # Y-axis label
plt.title("Time vs Distance")
plt.show()
```

```python
# 7. Create a Seaborn scatter plot using the tips dataset from Seaborn, with 'total_bill' on the x-axis and 'tip' on the y-axis.
tips = sns.load_dataset("tips")

sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day", style="time")
plt.title("Tips vs Total Bill")
plt.show()
```

```python
# 8. Generate a Seaborn heatmap for the correlation matrix of a DataFrame.
corr_matrix = tips.corr(numeric_only=True)  # get correlation of numeric columns

sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 9. Visualize the distribution of a numerical variable using Seaborn's distplot
sns.histplot(tips['total_bill'], kde=True, color='purple') # KDE is a smoothed curve that estimates the probability density of the data.
plt.title("Distribution of Total Bill")
plt.show()
```

```python
# 10. Create a Seaborn violin plot to visualize the distribution of a numerical variable across different categories.
sns.violinplot(data=tips, x="day", y="total_bill", palette="muted")
plt.title("Violin Plot of Total Bill by Day")
plt.show()
```

```python
# 11. Create a Seaborn box plot for the 'total_bill' column in the tips dataset, grouped by the 'day' column.
sns.boxplot(data=tips, x="day", y="total_bill", palette="Set2")
plt.title("Box Plot of Total Bill by Day")
plt.show()
```

```python
# 12. llustrate the count of categorical variables using Seaborn's countplot on a DataFrame column.
sns.countplot(data=tips, x="day", palette="pastel")
plt.title("Count of Records by Day")
plt.show()
```
