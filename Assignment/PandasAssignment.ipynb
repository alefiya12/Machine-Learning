```python
# 6. Difference between iloc and loc function, explain with example.
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Department': ['HR', 'Sales', 'Engineering']
}
df = pd.DataFrame(data, index=['a', 'b', 'c'])
df
```

The difference between .iloc[] and .loc[] in Pandas is based on how they access data in a DataFrame:

**.iloc[]** → Integer-location based indexing.
Uses row and column positions (0-based index).
It is used when you know the position of the element.
Syntax: **iloc[row_index, column_index]**

**.loc[]** → Label-based indexing.
Uses row labels and column names.
It is used when you know the names/labels of the element.
Syntax: **loc[row_label, column_name]**

```python
print("Row 0:\n",df.iloc[0],"\n")
print("Value at 2nd row and 1st column: ",df.iloc[1,0], "\n")
```

```python
print("Row with label 'b'\n", df.loc['b'], "\n")
print("Get the age of person with label 'c'", df.loc['c', 'Age'], "\n")
```
